```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Bar Plot

```python
fig, axis = plt.subplots(1,1)

data = {
    1:np.random.rand(10),
    2:np.random.rand(10),
    3:np.random.rand(10),
    4:np.random.rand(10)
}
df = pd.DataFrame(data)

sns.barplot(
    ax= axis,
    data=df,
    orient='h'
    )
```

Histogram & Distribution Plot

```python
fig, axis = plt.subplots(1,1)

data = {
    1:np.random.rand(10),
    2:np.random.rand(10),
    3:np.random.rand(10),
    4:np.random.rand(10)
}
df = pd.DataFrame(data)

sns.histplot(
    ax= axis,
    data=df
    )
sns.displot(
    data=df,
    kind='kde'
)
```

Scatter Plot

```python
fig, axis = plt.subplots(1,1)

data = {
    1:np.random.rand(10),
    2:np.random.rand(10),
    3:np.random.rand(10),
    4:np.random.rand(10)
}
df = pd.DataFrame(data)
# print(df.head())

sns.scatterplot(
    data=df
    )
```

```python
df1 = pd.DataFrame( {"Name": ['Apple', 'Apple', 'Apple', 'Orange', 'Orange', 'Orange', 'Banana', 'Banana', 'Banana'], 
                    "Location": ['India', 'USA', 'UK', 'India', 'USA', 'UK', 'India', 'USA', 'UK'],
                    "Cost":[80, 120, 140, 70, 110, 115, 60, 80, 95],
                    "Weight" : [1, 1, 1, 2, 2, 2, 3, 3, 3] }
)
df1.index.name = 'ROW_NAME'

# By default, the aggregate function is MEAN
df2 = df1.pivot_table(index='Location', columns='Name', values='Cost')
print(df2)

sns.pairplot(
    data=df2
    )
```

Categorical Plot

```python
data = {
    1:np.random.rand(5),
    2:np.random.rand(5),
    3:np.random.rand(5),
    4:np.random.rand(5)
}
df = pd.DataFrame(data)
# print(df.head())

sns.catplot(
    data=df,
    kind='bar',
    col=1,
    row=1
    )

plt.savefig("matplot_sns_figure.png")
```
