```python
import pandas as pd
import numpy as np

df1 = pd.DataFrame( {"Name": ['Apple', 'Apple', 'Apple', 'Orange', 'Orange', 'Banana', 'Banana', 'Mango', 'Grape', 'Apple', 'Apple', 'Orange', 'Mango', 'Banana'], 
                    "Location": ['India', 'USA', 'UK', 'India', 'USA', 'USA', 'UK', 'India', 'USA', 'India', 'USA', 'India', 'USA', 'USA'],
                    "Cost":[80, 120, 140, 70, 110, 50, 60, 90, 120, 85, 130, 80, 100, 70],
                    "Price": [80, 120, 140, 70, 110, 50, 60, 90, 120, 85, 130, 80, 100, 70]})

df1
```

SPLIT, APPLY, COMBINE



```python
# Group One Column by one Key
df1_new = df1['Cost'].groupby(df1['Name']).mean()
df1_new
```

```python
# Group One Column by multiple Key
df1_new = df1['Cost'].groupby([df1['Name'], df1['Location']]).mean()
df1_new
```

```python
# Group dataframe by one Key
df1_new = df1.groupby(df1['Name']).mean()
df1_new
```
