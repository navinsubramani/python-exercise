```python
import pandas as pd
import numpy as np
```

Apply Function : Column wise

```python
a = pd.DataFrame(np.arange(32).reshape(8, 4), index=['row1', 'row2', 'row3', 'row4', 'row5', 'row6', 'row7', 'row8'] ,columns=['col1', 'col2', 'col3', 'col4'])
f = lambda x: x.max()

print(a, a.apply(f), sep='\n\n')
```

Apply Function: Row wise

```python
print(a, a.apply(f, axis='columns'), sep='\n\n')

f = lambda x: pd.Series([x.max(), x.min()], index=['max', 'min'])
print(a.apply(f, axis='columns'), sep='\n\n')
```

Applymap, this works on each element of the datatframe

```python
a = pd.DataFrame(np.arange(32).reshape(8, 4), index=['row1', 'row2', 'row3', 'row4', 'row5', 'row6', 'row7', 'row8'] ,columns=['col1', 'col2', 'col3', 'col4'])
f = lambda x: x+10

print(a, a.applymap(f), sep='\n\n')
```

Map function, this works on series object

```python
a = pd.Series([0,1,2,3,4])
f = lambda x: x+10

print(a, a.map(f), sep='\n\n')
```
