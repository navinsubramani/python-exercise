```python
import pandas as pd

json_file_path = "pandas_example_dataset_2.json"
hdf5_file_path = "pandas_write_tohdf5"
excel_file_path = "pandas_write_toexcel.xlsx"

df = pd.read_json(json_file_path)
df
```

HDF5



HDF5 Operations - Method 1

```python
# Write
df.to_hdf(hdf5_file_path, 'obj1', format='table')

# Read
a = pd.read_hdf(hdf5_file_path, 'obj1')
a
```

HDF5 Operations - Method 2

```python
hdfStore = pd.HDFStore(hdf5_file_path)

# Write
hdfStore['obj1'] = df

# Read
a = hdfStore['obj1']
a
```

HDF5 Operation - Method 3

```python
hdfStore = pd.HDFStore(hdf5_file_path)

# Write
hdfStore.put('obj1', df, format='table')

# Read
a = hdfStore.select('obj1')
```

EXCEL



Excel Operation

```python
# Write to a Excel File
print(df)
excel_write_reference = pd.ExcelWriter(excel_file_path) # Create a reference
df.to_excel(excel_write_reference) # Write to the excel file
excel_write_reference.save()
excel_write_reference.close()

# Read from a Excel File

excel_read_reference = pd.ExcelFile(excel_file_path)
a = pd.read_excel(excel_read_reference)
excel_read_reference.close()

print(a)
```
