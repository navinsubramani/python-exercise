```python
import pandas as pd
```

Min, Max, Standard Deviation (by columns & by rows)

```python
dict1 = {
    'a':[4,1,1,2],
    'b':[2,9,5,6],
    'c':[5,3,4,1]
}
a = pd.DataFrame(dict1, index=[1,2,3,4])

print(a, a.min(), a.max(), a.std(), sep='\n\n')
```

Index of Min, Max (by columns & by rows)

```python
dict1 = {
    'a':[4,1,1,2],
    'b':[2,9,5,6],
    'c':[5,3,4,1]
}
a = pd.DataFrame(dict1, index=['first','second','third','fourth'])

print(a, a.idxmax(), a.idxmin(), sep='\n\n')
```

Describe for Numeric data

```python
dict1 = {
    'a':[4,1,1,2],
    'b':[2,9,5,6],
    'c':[5,3,4,1]
}
a = pd.DataFrame(dict1, index=['first','second','third','fourth'])

print(a, a.describe(), sep='\n\n')
```

Describe for non numeric data

```python
dict1 = {
    'a':['a', 'b', 'a', 'c'],
    'b':[2,9,5,6],
    'c':[5,3,4,1]
}
a = pd.DataFrame(dict1, index=['first','second','third','fourth'])

print(a, a['a'].describe(), sep='\n\n')
```

Correlation & Covariance

```python
dict1 = {
    'a':[4,1,1,2],
    'b':[2,9,5,6],
    'c':[5,3,4,1],
    'd':[2,4,5,2]
}
a = pd.DataFrame(dict1, index=['first','second','third','fourth'])

print(a, a.corr(), a['a'].corr(a['b']), a.cov(), sep='\n\n')
```
