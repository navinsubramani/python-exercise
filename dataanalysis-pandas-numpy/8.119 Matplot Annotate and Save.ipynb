```python
import matplotlib.pyplot as plt
import numpy as np
```

Annotate & Save the figure

```python
fig, axis = plt.subplots(1,1)

plot_labels = {'title': 'Sample Plot', 'xlabel':'X-axis', 'ylabel':'Y-axis'}
data = np.random.rand(50)

axis.set(**plot_labels)
axis.plot(data.cumsum(), 'go--', label='Data-1')

# Annotation
axis.annotate('Note', xy=[20, 10], xytext=(20,20), arrowprops={'facecolor':'red', 'headwidth':10, 'width':5, 'headlength':4}, horizontalalignment='left', verticalalignment='top')
axis.legend(loc='best')

# Save the image
fig.savefig('matplot_figure.png')
```
