Pandas Dataframe: Row Indexing

```python
data =  {
        'Subject':{'one': 'English', 'two':'Math', 'three':'Science'},
        'Marks':{'one': 80, 'two':90},
        'Teacher':{'three':'Adam', 'two':'Lina'}
        }

df = pd.DataFrame(data, columns=['Subject', 'Teacher', 'Marks'])
df
```

```python
# Indexing using the names

print(df.loc['one'], df.loc['one':'two'], sep="\n", end="\n\n")

# Positional Indexing

print(df.iloc[1])
```

```python
# Index column & rows

print(df.loc['one', 'Marks'])
```

We can use "AT & iAT" instead of "LOC & iLOC". But "AT & iAT" is used only on the scalar values

Heirarchiical Indexing

```python
print(df)

a = df.loc[:"two"][df.Marks > 85]
print(a)
```

Summary:

Negative Indexing doesn't work in Pandas. If we need to use negative indexing, use "iLOC" to index with negative numbers.
